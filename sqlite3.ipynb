```python
import sqlite3
import pandas as pd
```

```python
# path = "../input/" 
# database = path + 'spotify.sqlite'
database = 'spotify.sqlite'

conn = sqlite3.connect(database)
# Additional for solve error Could not decode to UTF-8 column (SQLITE)
conn.text_factory = lambda b: b.decode(errors = 'ignore')

tables = pd.read_sql("""SELECT *
                        FROM sqlite_master
                        WHERE type='table';""", conn)
print("Conection SuccessFull",conn)
```

#### Select Data

```python
# Bentuk pertama
pd.read_sql_query("SELECT * FROM albums", conn)
```

```python
# Bentuk kedua
pd.read_sql("SELECT * FROM albums", conn)
```

Dalam kesimpulannya, perbedaan utama antara pd.read_sql_query() dan pd.read_sql() adalah bahwa pd.read_sql_query() digunakan khusus untuk mengeksekusi pernyataan SQL yang mengembalikan hasil dan pd.read_sql() dapat digunakan untuk mengeksekusi pernyataan SQL yang mengembalikan hasil dan yang tidak mengembalikan hasil.

#### Limit Data and Order By

```python
pd.read_sql_query("SELECT * FROM artists ORDER BY followers desc limit 50",conn)
```

#### Conditional Where

```python
pd.read_sql_query("SELECT * from artists WHERE popularity > 90 limit 10",conn)
```

#### Count Data

```python
pd.read_sql_query("SELECT COUNT(id) FROM artists ",conn)
```

#### Aggregate Functions(SUM,AVG,MIN,MAX)

```python
pd.read_sql_query(' ' ' SELECT AVG(popularity) AS "Average popularity" FROM artists;  ' ' ' ,conn )
```
